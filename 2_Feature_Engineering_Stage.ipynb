```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
#lets load our previous work, the preprocessed dataset
processed_file = "titanic_preprocessed.csv"
df = pd.read_csv(preprocessed_file)
print(f"\nProcessed data saved to {preprocessed_file}")
```

```python
print("- The data before feature engineering!")
print(df.head())
print("\n")
df.info()
print("\n")
```

```python
# after running 'df.head' and 'df.info' and you find the 6 columns  we were working on is clean, then congrats you dont have any problems!
# lets start with feature engineering(Scaling, Fitting, transforming, and creating interaction features)
# Selecting and Scaling numerical features(age and Fare)
scaler = StandardScaler()
columns_to_scale = ["age", "fare"]
print (f"Scaling columns: {columns_to_scale}")
print(df[columns_to_scale].dtypes)
print(df[columns_to_scale].isnull().sum())
print(columns_to_scale)
print(df.dtypes)
```

```python
# Fitting and Transforming what we selected
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
#scaler.fit_transform(df[['age', 'fare']]) 
```

```python
# Creating Interaction Feature on 'pclass' and 'sex'
# This might capture if the effect of class differs by sex
df["pclass_sex_interaction"] = df["pclass"] * df["sex"]
print("interaction feature created: pclass_sex_interaction")
```

```python
# Same as before
print("\n -Data After Feature Engineering")
print(df.head())
print("\n")
df.info()
```

```python
engineered_file = "titanic_engineered.csv"
df.to_csv(engineered_file, index=False)
print(f"\nFeature engineered data saved to {engineered_file}")
```

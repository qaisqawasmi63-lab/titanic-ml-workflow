```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Loading our previous work from feature engineering

engineered_file = "titanic_engineered.csv"
df = pd.read_csv(engineered_file)
#print(f"\nFeature engineered data saved to {engineered_file}")
```

```python
print(" -Data for clustering analysis")
print(df.head())
print("\n")
```

```python
# Selecting
features_for_clustering = ["age", "fare", "pclass"]
print(f"Features selected for clustering: {features_for_clustering}")
X_cluster = df[features_for_clustering]
```

```python
#from sklearn.cluster import KMeans
print("\n -Determining Optimal K(elbow method)")
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state = 42, n_init = 10)
    kmeans.fit(X_cluster)
    inertia.append(kmeans.inertia_)
   
```

```python
# plotting the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker ='o')
plt.xlabel("number of clusters(k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
elbow_plot_file = "elbow_plot.png"
plt.savefig(elbow_plot_file)
#plt.show() # use this if you want to see the graph below
print(f"Elbow plot saved to {elbow_plot_file}")
plt.close() # to prevent displaying in output log you have to close the plot
k_optimal = 4 # here we have to choose a value for k i assumed k=4 and it is reasonable
print(f"\nSelected optimal k = {k_optimal}")
```

```python
# Applying K-Means with the chosen k
kmeans = KMeans(n_clusters = k_optimal, random_state = 42, n_init = 10)
cluster_labels =kmeans.fit_predict(X_cluster)
df["cluster"] = cluster_labels
print("\n -Cluster Analysis Results")
print(f"Cluster labels assigned (first 10): {cluster_labels[:10]}")
```

```python
#analyzing cluster char
print("\nCluster Centers(Mean values of features for each cluster):")
cluster_centers =pd.DataFrame(kmeans.cluster_centers_, columns = features_for_clustering)
print(cluster_centers)
print("\nCluster Sizes:")
print(df["cluster"].value_counts())
```

```python
# Here we have to upload the preprocessed data to get the original scales for interpretation to analyze char by cluster
preprocessed_file = "titanic_preprocessed.csv"
df_preprocessed = pd.read_csv(preprocessed_file)
#adding cluster labels
df_preprocessed["cluster"] = cluster_labels
print("\nMean values of key features per cluster")
cluster_anaylsis = df_preprocessed.groupby("cluster")[["pclass", "age", "fare", "survival", "sex"]].mean()
print(cluster_anaylsis)
```

```python
# saving results
cluster_data_file = "titanic_clustered.csv"
df.to_csv(cluster_data_file, index = False)
print(f"\nData with cluster labels saved to {cluster_data_file}")
```

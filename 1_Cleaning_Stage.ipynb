```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset

df = pd.read_csv("titanic.csv")
print('-Original Data Info')
df.info()
print('\n  -Original Head')
print(df.head())
print("\n")
```

```python
# Data cleaning
# You can start with any tag, Im going to start with age.
# replacing Comas!
df['age'] = df['age'].astype(str).str.replace(',','.', regex = False)
df['age'] = pd.to_numeric(df['age'], errors = 'coerce')
```

```python
# Updating empty or missing 'age' values with median.
age_median = df['age'].median()
print(f"Updating missing or empty 'age' values with median: {age_median:.2f}")
df['age'].fillna(age_median, inplace = True)
```

```python
# For the second step, im going to work on fare(cleaning), replacing comas and converting to numeric.
df['fare'] = df['fare'].astype(str).str.replace(',','.', regex = False)
df['fare'] = pd.to_numeric(df['fare'], errors = 'coerce')
```

```python
#checking NaNs if it was introduced in 'fare', I use 'if statement' to me its simple and easy.
if df['fare'].isnull().sum() > 0:
    # I like my work's interface to show me easy user understanding, so im doing this message.
    print("NaNs found in 'fare', imputing with median.")
    fare_median = df['fare'].median()
    df['fare'].fillna(fare_median, inplace = True)
```

```python
# Converting 'Pclass'
pclass_mapping = {'1st': 1, '2nd': 2, '3rd': 3}
df['pclass'] = df['pclass'].map(pclass_mapping)
if df['pclass'].isnull().sum() > 0:
 print("Unmapped values found in 'pclass")
#if it requires handling i'll add them as a comment.
#pclass_mode = df['pclass'].mode()[0]
#df['pclass'].fillna(pclass_mode, inplace = True)
```

```python
# converting 'sex'
sex_mapping ={'female': 0, 'male':1}
df['sex'] = df['sex'].map(sex_mapping)
```

```python
# converting 'embarked'
# because embarked have 3 categories im doing one hot-code for it that satisfies the requirement
df = pd.get_dummies(df, columns=['embarked'], prefix='embarked', drop_first=True) # drop_first=True to avoid multicollinearity(sometimes it happen :) )

```

```python
# showing processed data info!
print("\n -Processed Data Info")
df.info()
print("\n -Processed Head")
print(df.head())
```

```python
processed_file ="titanic_preprocessed.csv"
df.to_csv(processed_file, index = False)
print(f"\nProcessed data saved to {processed_file}")
```
